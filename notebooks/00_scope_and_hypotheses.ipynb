# Scope and Hypotheses


## Goal
Freeze the exact project question and success criteria before coding deeper.

## 1. Final Dataset Slice
- **Amazon Reviews 2023**
- **4 Categories**: Electronics, Home and Kitchen, Beauty and Personal Care, Sports and Outdoors
- **Sample Size**: 500,000 per category (2,000,000 total reviews)
- **Split Strategy**: By `parent_asin` to prevent data leakage.

## 2. Primary Tasks
1. **Sentiment/Rating Classification** (3-class: negative, neutral, positive)
2. **Category Classification** (4-class: category prediction)
3. **Retrieval**: Nearest-neighbor search for similar reviews.
4. **Clustering**: Analyzing topical clusters in latent space.

## 3. Main Hypotheses
- **Sparse Baselines (TF-IDF, LSA)** remain highly competitive for direct lexical tasks like classification.
- **Dense Learned Embeddings (Word2Vec)** will provide significantly better geometry and semantic neighborhood coherence for retrieval and clustering tasks.

## 4. Strong Result Criteria
A successful project will not just report classification accuracy. It will use UMAP and nearest-neighbor purity to visually and mathematically demonstrate *why* the continuous dense space aligns better with human perception of review similarity than sparse vectors.
